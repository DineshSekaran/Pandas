# Group By in Pandas

```python
drinks=pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head(2)
```

```python
drinks.beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks[drinks.continent=='Africa']
```

```python
drinks[drinks.continent=='Africa'].beer_servings.mean()
```

```python
drinks.groupby('continent').beer_servings.max()
```

```python
drinks.groupby('continent').beer_servings.agg(['mean','min','max','count'])
```

```python
# ALl numeric column mean
drinks.groupby('continent').mean()
```

```python
%matplotlib inline
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```

# Handle missing values in pandas

```python
ufo=pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.tail(8)
```

```python
ufo.fillna('It is Null').head(19)
```

```python
ufo.isnull()
```

```python
ufo[ufo.notnull()]
```

```python
ufo.isnull().sum()
```

```python
# Subset of Data Frame
ufo[ufo.City.isnull()]
```

```python
ufo.shape
```

```python
ufo.dropna(how='any').shape # any of rows are null
```

```python
ufo.dropna(how='all').shape # only drop allvaues are null
```

```python
# drop rows either of column values are null
ufo.dropna(subset=['City','Shape Reported'],how='any').shape
```

```python
# drop 4 rows having both columns are null
ufo.dropna(subset=['City','Shape Reported'],how='all').shape
```

```python
ufo['Shape Reported'].value_counts()
```
