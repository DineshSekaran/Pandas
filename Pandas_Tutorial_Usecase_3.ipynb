# Kaggle Data

```python
train=pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head(4)
```

```python
feature_Class=['Pclass','Parch']
```

```python
x=train.loc[:,feature_Class]
```

```python
x.shape
```

```python
y=train.Survived
```

```python
y.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
logreg.fit(x,y)
```

```python
test=pd.read_csv('http://bit.ly/kaggletest')
```

```python
test.head(4)
```

```python
x_new=test.loc[:,feature_Class]
```

```python
x_new.shape
```

```python
new_pred_class=logreg.predict(x_new)
```

```python
new_pred_class
```

```python
# Created csv file in my working directory 
pd.DataFrame({'Passengerid':test.PassengerId,'Survived':new_pred_class}).to_csv('sub.csv')
```
